```python
"""
Final Machine Learning Script for HFT LOB Forecasting

This script implements both regression and classification tasks for HFT LOB data.
Multiple models (MLP, LSTM, CNN, Dual-Task, and an autoencoder-based classifier)
are trained using time-series splits. For each fold, the script runs the regression 
and classification experiments for each model type simultaneously and displays the results:
    - Regression results: RMSE, MAE, and a plot (first 1000 time steps) with x-axis "Time Step",
      y-axis "Mid-Price", and the actual mid-price labeled as "Actual Mid Prices".
    - Classification results: Overall accuracy plus a full classification report.
At the end, a summary of the average regression and classification metrics is printed and additional plots are generated:
    * A combined plot of predicted mid‑prices from all regression models versus actual mid‑prices.
    * A plot of the ensemble regression prediction (average of all regression models).
    * The confusion matrix of the best classification model (based on average accuracy, using all models).
    * An ensemble classification output (via majority voting) with its classification report and confusion matrix.
    
The script handles missing, NaN, and infinite values.
"""

import sys
import hdf5storage
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, confusion_matrix
from keras.models import Sequential, Model
from keras.layers import Input, Dense, LSTM, Conv1D, MaxPooling1D, Flatten, Dropout
from keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, accuracy_score, precision_recall_fscore_support
from imblearn.over_sampling import ADASYN
from keras.utils import to_categorical, Sequence
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
import random
import seaborn as sns
import warnings
from sklearn.exceptions import UndefinedMetricWarning
from scipy.stats import mode  # Added for majority voting ensemble

# Suppress warnings regarding undefined metrics
warnings.filterwarnings("ignore", category=UndefinedMetricWarning)

# =============================================================================
# Custom Loss and Data Generator Definitions
# =============================================================================

def weighted_focal_loss(alpha, gamma=2.):
    """
    Returns a weighted focal loss function for handling class imbalance.
    
    Args:
        alpha (list): List of weights for each class.
        gamma (float): Focusing parameter.
    
    Returns:
        function: Loss function.
    """
    alpha = tf.constant(alpha, dtype=tf.float32)
    
    def loss(y_true, y_pred):
        y_pred = tf.clip_by_value(y_pred, tf.keras.backend.epsilon(), 1. - tf.keras.backend.epsilon())
        cross_entropy = -y_true * tf.math.log(y_pred)
        weight = y_true * alpha
        focal_loss = weight * tf.pow(1 - y_pred, gamma) * cross_entropy
        return tf.reduce_sum(focal_loss, axis=1)
    
    return loss

class StratifiedBatchGenerator(Sequence):
    """
    Generates stratified batches with on-the-fly Gaussian noise augmentation for minority class samples.
    
    Args:
        X (np.array): Input features.
        y (np.array): One-hot encoded labels.
        batch_size (int): Number of samples per batch.
        noise_level (float): Standard deviation for Gaussian noise.
    """
    def __init__(self, X, y, batch_size=256, noise_level=0.01, **kwargs):
        """
        Initializes the batch generator.
        
        Args:
            X (np.array): Input features.
            y (np.array): One-hot encoded labels.
            batch_size (int): Batch size.
            noise_level (float): Noise level for augmentation.
            **kwargs: Additional keyword arguments.
        """
        super().__init__(**kwargs)
        self.X = X
        self.y = y
        self.batch_size = batch_size
        self.noise_level = noise_level
        self.num_samples = X.shape[0]
        
    def __len__(self):
        """
        Returns the number of batches per epoch.
        
        Returns:
            int: Number of batches.
        """
        return int(np.ceil(self.num_samples / self.batch_size))
    
    def __getitem__(self, idx):
        """
        Generates one batch of data.
        
        Args:
            idx (int): Batch index.
        
        Returns:
            tuple: (Batch of input features, corresponding one-hot encoded labels)
        """
        batch_indices = np.random.choice(self.num_samples, size=self.batch_size, replace=True)
        X_batch = self.X[batch_indices].copy()
        y_batch = self.y[batch_indices]
        minority_indices = np.where(np.argmax(y_batch, axis=1) != 1)[0]
        if minority_indices.size > 0:
            noise = np.random.normal(0, self.noise_level, size=X_batch[minority_indices].shape)
            X_batch[minority_indices] += noise
        return X_batch, y_batch

# =============================================================================
# Reproducibility Setup
# =============================================================================
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

# =============================================================================
# Data Preparation & Feature Engineering
# =============================================================================
dataset_path = '/Users/ridhimagupta/Downloads/S092215-v50-AMZN_OCT2_states.mat'
lob_input = hdf5storage.loadmat(dataset_path)
lob_amazon = pd.DataFrame.from_dict(lob_input['LOB'])

# Set column names for the LOB data
lob_amazon.columns = [
    "Time", "Mid_Price", "Spread", 
    "AskPrice1", "AskVolume1", "BidPrice1", "BidVolume1", 
    "AskPrice2", "AskVolume2", "BidPrice2", "BidVolume2", 
    "AskPrice3", "AskVolume3", "BidPrice3", "BidVolume3", 
    "AskPrice4", "AskVolume4", "BidPrice4", "BidVolume4", 
    "AskPrice5", "AskVolume5", "BidPrice5", "BidVolume5", 
    "AskPrice6", "AskVolume6", "BidPrice6", "BidVolume6", 
    "AskPrice7", "AskVolume7", "BidPrice7", "BidVolume7", 
    "AskPrice8", "AskVolume8", "BidPrice8", "BidVolume8", 
    "AskPrice9", "AskVolume9", "BidPrice9", "BidVolume9", 
    "AskPrice10", "AskVolume10", "BidPrice10", "BidVolume10"
]
print("Dataset shape:", lob_amazon.shape)

# Select a subset of the data
lob_amazon_subset = lob_amazon.iloc[250000:500000]
print("Subset shape:", lob_amazon_subset.shape)

# Create additional features: lag, rolling mean, and EMA for selected columns
features_for_engineering = lob_amazon_subset.copy()
look_back = 10
for col in ['AskPrice1', 'BidPrice1', 'Mid_Price']:
    features_for_engineering.loc[:, f'{col}_lag'] = features_for_engineering[col].shift(look_back)
    features_for_engineering.loc[:, f'{col}_rolling_mean'] = features_for_engineering[col].rolling(window=look_back).mean()
    features_for_engineering.loc[:, f'{col}_ema'] = features_for_engineering[col].ewm(span=look_back).mean()

# Handle infinite values and drop rows with missing values
features_for_engineering.replace([np.inf, -np.inf], np.nan, inplace=True)
features_for_engineering = features_for_engineering.dropna()

# Define feature matrix (inputs) and regression target (Mid_Price)
feature_matrix = features_for_engineering.drop(['Time', 'Mid_Price', 'Spread'], axis=1)
mid_price_series = lob_amazon_subset['Mid_Price']

# Normalize features and target
scaler_features = MinMaxScaler()
feature_matrix_scaled = scaler_features.fit_transform(feature_matrix)
scaler_target = MinMaxScaler()
mid_price_series_scaled = scaler_target.fit_transform(mid_price_series.values.reshape(-1, 1))

def create_classification_labels(mid_prices, threshold=0.0):
    """
    Generate classification labels based on price differences.
    
    Args:
        mid_prices (array-like): Array of mid-price values.
        threshold (float, optional): Threshold for determining stationary movement.
    
    Returns:
        np.array: Labels (0: decrease, 1: stationary, 2: increase).
    """
    mid_prices = np.ravel(mid_prices)
    diff = np.diff(mid_prices, prepend=mid_prices[0])
    labels = np.where(diff > threshold, 2, np.where(diff < -threshold, 0, 1))
    return labels

classification_labels = create_classification_labels(features_for_engineering['Mid_Price'].values)
tscv_reg = TimeSeriesSplit(n_splits=5)

def calculate_regression_metrics(y_true, y_pred):
    """
    Calculate RMSE and MAE for regression predictions.
    
    Args:
        y_true (np.array): True target values.
        y_pred (np.array): Predicted target values.
    
    Returns:
        tuple: (RMSE, MAE)
    """
    y_true_orig = scaler_target.inverse_transform(y_true.reshape(-1, 1))
    y_pred_orig = scaler_target.inverse_transform(y_pred.reshape(-1, 1))
    rmse = np.sqrt(np.mean((y_true_orig - y_pred_orig)**2))
    mae = mean_absolute_error(y_true_orig, y_pred_orig)
    return rmse, mae

# =============================================================================
# Model Creation Functions for Regression and Classification
# =============================================================================

def create_mlp_model(input_shape, depth='shallow'):
    """
    Create an MLP regression model.
    
    Args:
        input_shape (tuple): Shape of input features.
        depth (str): Model depth ('shallow', 'deep', or 'very deep').
    
    Returns:
        keras.Model: Compiled regression model.
    """
    model = Sequential([Input(shape=(input_shape[1],))])
    if depth == 'shallow':
        model.add(Dense(100, activation='relu'))
    elif depth == 'deep':
        model.add(Dense(100, activation='relu'))
        model.add(Dense(32, activation='relu'))
    else:
        model.add(Dense(100, activation='relu'))
        model.add(Dense(100, activation='relu'))
        model.add(Dense(100, activation='relu'))
        model.add(Dense(32, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer=Adam(), loss='mse', metrics=['mse'])
    return model

def create_lstm_model(input_shape, look_back=10):
    """
    Create an LSTM regression model.
    
    Args:
        input_shape (tuple): Shape of input features.
        look_back (int): Number of time steps for the LSTM.
    
    Returns:
        keras.Model: Compiled LSTM regression model.
    """
    model = Sequential([
        Input(shape=(look_back, input_shape[1])),
        LSTM(256, return_sequences=False),
        Dense(1)
    ])
    model.compile(optimizer=Adam(), loss='mse', metrics=['mse'])
    return model

def create_cnn_model(input_shape, look_back=10):
    """
    Create a CNN regression model.
    
    Args:
        input_shape (tuple): Shape of input features.
        look_back (int): Number of time steps for the CNN.
    
    Returns:
        keras.Model: Compiled CNN regression model.
    """
    model = Sequential([
        Input(shape=(look_back, input_shape[1])),
        Conv1D(64, kernel_size=3, activation='relu'),
        Conv1D(64, kernel_size=3, activation='relu'),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(100, activation='relu'),
        Dense(1, activation='linear')
    ])
    model.compile(optimizer=Adam(), loss='mse', metrics=['mse'])
    return model

def create_mlp_classification_model(input_shape, depth='shallow'):
    """
    Create an MLP classification model.
    
    Args:
        input_shape (tuple): Shape of input features.
        depth (str): Model depth ('shallow', 'deep', or 'very deep').
    
    Returns:
        keras.Model: Compiled MLP classification model.
    """
    updated_alpha = [0.7, 0.1, 0.7]
    model = Sequential([Input(shape=(input_shape[1],))])
    if depth == 'shallow':
        model.add(Dense(100, activation='relu'))
        model.add(Dropout(0.4))
    elif depth == 'deep':
        model.add(Dense(100, activation='relu'))
        model.add(Dropout(0.4))
        model.add(Dense(32, activation='relu'))
        model.add(Dropout(0.4))
    else:
        model.add(Dense(100, activation='relu'))
        model.add(Dropout(0.4))
        model.add(Dense(100, activation='relu'))
        model.add(Dropout(0.4))
        model.add(Dense(100, activation='relu'))
        model.add(Dropout(0.4))
        model.add(Dense(32, activation='relu'))
        model.add(Dropout(0.4))
    model.add(Dense(3, activation='softmax'))
    model.compile(optimizer=Adam(), loss=weighted_focal_loss(updated_alpha, gamma=2.), metrics=['accuracy'])
    return model

def create_lstm_classification_model(input_shape, look_back=10):
    """
    Create an LSTM classification model.
    
    Args:
        input_shape (tuple): Shape of input features.
        look_back (int): Number of time steps for the LSTM.
    
    Returns:
        keras.Model: Compiled LSTM classification model.
    """
    updated_alpha = [0.7, 0.1, 0.7]
    model = Sequential([
        Input(shape=(look_back, input_shape[1])),
        LSTM(256, return_sequences=False),
        Dropout(0.4),
        Dense(3, activation='softmax')
    ])
    model.compile(optimizer=Adam(), loss=weighted_focal_loss(updated_alpha, gamma=2.), metrics=['accuracy'])
    return model

def create_cnn_classification_model(input_shape, look_back=10):
    """
    Create a CNN classification model.
    
    Args:
        input_shape (tuple): Shape of input features.
        look_back (int): Number of time steps for the CNN.
    
    Returns:
        keras.Model: Compiled CNN classification model.
    """
    updated_alpha = [0.7, 0.1, 0.7]
    model = Sequential([
        Input(shape=(look_back, input_shape[1])),
        Conv1D(64, kernel_size=3, activation='relu'),
        Dropout(0.4),
        Conv1D(64, kernel_size=3, activation='relu'),
        Dropout(0.4),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(100, activation='relu'),
        Dropout(0.4),
        Dense(3, activation='softmax')
    ])
    model.compile(optimizer=Adam(), loss=weighted_focal_loss(updated_alpha, gamma=2.), metrics=['accuracy'])
    return model

def create_autoencoder_classification_model(input_shape, encoding_dim=32):
    """
    Create an autoencoder-based classification model.
    
    Args:
        input_shape (tuple): Shape of input features.
        encoding_dim (int): Dimension of the encoding layer.
    
    Returns:
        tuple: (Full classification model, autoencoder model, encoder model)
    """
    updated_alpha = [0.7, 0.1, 0.7]
    input_layer = Input(shape=(input_shape[1],))
    encoded = Dense(encoding_dim, activation='relu')(input_layer)
    decoded = Dense(input_shape[1], activation='linear')(encoded)
    autoencoder = Model(input_layer, decoded)
    encoder = Model(input_layer, encoded)
    autoencoder.compile(optimizer=Adam(), loss='mse')
    
    encoded_input = Input(shape=(encoding_dim,))
    classifier_output = Dense(3, activation='softmax')(encoded_input)
    classifier = Model(encoded_input, classifier_output)
    
    input_auto = Input(shape=(input_shape[1],))
    encoded_repr = encoder(input_auto)
    classification_output = classifier(encoded_repr)
    full_model = Model(input_auto, classification_output)
    full_model.compile(optimizer=Adam(), loss=weighted_focal_loss(updated_alpha, gamma=2.), metrics=['accuracy'])
    return full_model, autoencoder, encoder

def create_dual_task_model(input_shape, look_back=10, encoding_dim=64):
    """
    Create a dual-task model that performs both regression and classification.
    
    Args:
        input_shape (tuple): Shape of input features.
        look_back (int): Number of time steps.
        encoding_dim (int): Dimension of the shared encoding.
    
    Returns:
        keras.Model: Compiled dual-task model.
    """
    updated_alpha = [0.7, 0.1, 0.7]
    inputs = Input(shape=(look_back, input_shape[1]))
    x = LSTM(256, return_sequences=False)(inputs)
    x = Dropout(0.4)(x)
    shared = Dense(128, activation='relu')(x)
    shared = Dropout(0.4)(shared)
    reg_output = Dense(1, activation='linear', name='regression')(shared)
    class_output = Dense(3, activation='softmax', name='classification')(shared)
    model = Model(inputs=inputs, outputs=[reg_output, class_output])
    model.compile(optimizer=Adam(),
                  loss={'regression': 'mse', 'classification': weighted_focal_loss(updated_alpha, gamma=2.)},
                  metrics={'regression': ['mse'], 'classification': ['accuracy']})
    return model

# =============================================================================
# RUN MODELS SIMULTANEOUSLY PER MODEL TYPE PER FOLD
# =============================================================================

def run_models_simultaneously():
    """
    Run regression and classification experiments simultaneously for each model type in each fold.
    For each model type (MLP Shallow, MLP Deep, MLP Very Deep, LSTM, CNN, Dual-Task),
    this function:
        - Runs a regression experiment (prints RMSE/MAE and plots prediction for first 1000 time steps).
        - Runs a classification experiment (using resampled data for non-sequential models),
          printing overall accuracy and the full classification report.
    At the end, the function:
        - Plots a combined graph of predicted mid-prices for all regression models versus actual mid-prices.
        - Plots the ensemble regression prediction (average of all regression models).
        - Determines the best classification model (based on average accuracy, using all models) and plots its confusion matrix.
        - Computes an ensemble classification prediction using majority voting (hard voting) across all models,
          prints its classification report, and plots its confusion matrix.
    """
    fold_no = 1
    reg_summary = {}   # {model_name: list of (rmse, mae)}
    class_summary = {} # {model_name: list of accuracy}
    best_lstm_acc = 0
    best_lstm_cm = None
    final_reg_preds = {}       # Final regression predictions for each model type from the last fold
    final_class_preds = {}     # Final predicted probabilities for classification for each model type
    final_class_cm = {}        # Final confusion matrices for classification models
    
    for name in ['MLP Shallow', 'MLP Deep', 'MLP Very Deep', 'LSTM', 'CNN', 'Dual-Task']:
        reg_summary[name] = []
    for name in ['MLP Shallow', 'MLP Deep', 'MLP Very Deep', 'LSTM', 'CNN', 'Autoencoder', 'Dual-Task']:
        class_summary[name] = []
    
    # Dictionary to store macro-averaged classification metrics for each model
    class_metrics = {}
    def compute_metrics(y_true, y_pred, model_name):
        """
        Compute macro-averaged precision, recall, and F1-score for a classifier
        and store the results in the `class_metrics` dictionary.

        Args:
        y_true (array-like): Ground truth class labels.
        y_pred (array-like): Predicted class labels.
        model_name (str): Name of the model (used as key for storing results).
    
        Stores:
        class_metrics[model_name]: Tuple containing (precision, recall, f1_score).
        """
        # Calculate macro-averaged metrics to reflect class-wise balance
        precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='macro')
        class_metrics[model_name] = (precision, recall, f1)
    
    # Iterate over folds
    for train_index, test_index in tscv_reg.split(feature_matrix_scaled):
        print(f"\n=== Fold {fold_no} ===")
        X_train, X_test = feature_matrix_scaled[train_index], feature_matrix_scaled[test_index]
        y_train, y_test = mid_price_series_scaled[train_index], mid_price_series_scaled[test_index]
        y_train_class = classification_labels[train_index]
        y_test_class = classification_labels[test_index]
        
        X_train_lstm = np.array([X_train[i:i+look_back] for i in range(len(X_train)-look_back)])
        y_train_lstm = y_train[look_back:]
        X_test_lstm = np.array([X_test[i:i+look_back] for i in range(len(X_test)-look_back)])
        y_test_lstm = y_test[look_back:]
        y_train_class_lstm = y_train_class[look_back:]
        y_test_class_lstm = y_test_class[look_back:]
        
        early_stop = EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)
        
        # ---- MLP Models (Shallow, Deep, Very Deep) ----
        for depth, label in zip(['shallow', 'deep', 'very deep'], ['MLP Shallow', 'MLP Deep', 'MLP Very Deep']):
            print(f"\n--- {label} ---")
            model_reg = create_mlp_model(X_train.shape, depth)
            model_reg.fit(X_train, y_train, epochs=10, batch_size=1024,
                          validation_data=(X_test, y_test), callbacks=[early_stop], verbose=0)
            pred_reg = model_reg.predict(X_test)
            rmse, mae = calculate_regression_metrics(y_test, pred_reg)
            reg_summary[label].append((rmse, mae))
            final_reg_preds[label] = pred_reg
            print(f"{label} Regression - RMSE: {rmse:.4f}, MAE: {mae:.4f}")
            plt.figure(figsize=(12,6))
            plt.plot(scaler_target.inverse_transform(y_test)[:1000], label='Actual Mid Prices')
            plt.plot(scaler_target.inverse_transform(pred_reg)[:1000], label=f'{label} Prediction')
            plt.xlabel("Time Step")
            plt.ylabel("Mid-Price")
            plt.title(f"Fold {fold_no} - {label} Regression")
            plt.legend()
            plt.show()
            
            adasyn = ADASYN(random_state=42)
            X_train_bal, y_train_bal = adasyn.fit_resample(X_train, y_train_class)
            classes = np.unique(y_train_bal)
            class_weights = compute_class_weight('balanced', classes=classes, y=y_train_bal)
            class_weight_dict = {i: weight for i, weight in zip(classes, class_weights)}
            y_train_bal_cat = to_categorical(y_train_bal, num_classes=3)
            y_test_class_cat = to_categorical(y_test_class, num_classes=3)
            model_cls = create_mlp_classification_model(X_train.shape, depth)
            model_cls.fit(X_train_bal, y_train_bal_cat, epochs=50, batch_size=256,
                          validation_data=(X_test, y_test_class_cat), callbacks=[early_stop], verbose=0,
                          class_weight=class_weight_dict)
            pred_cls = model_cls.predict(X_test)
            final_class_preds[label] = pred_cls
            y_pred_cls = np.argmax(pred_cls, axis=1)
            acc = accuracy_score(y_test_class, y_pred_cls)
            class_summary[label].append(acc)
            print(f"{label} Classification - Accuracy: {acc:.4f}")
            print(classification_report(y_test_class, y_pred_cls, zero_division=0))
            compute_metrics(y_test_class, y_pred_cls, label)
            final_class_cm[label] = confusion_matrix(y_test_class, y_pred_cls)
        
        # ---- LSTM Model ----
        print("\n--- LSTM ---")
        model_reg = create_lstm_model(X_train.shape)
        model_reg.fit(X_train_lstm, y_train_lstm, epochs=10, batch_size=1024,
                      validation_data=(X_test_lstm, y_test_lstm), callbacks=[early_stop], verbose=0)
        pred_reg = model_reg.predict(X_test_lstm)
        pred_reg = np.pad(pred_reg, ((look_back,0),(0,0)), mode='edge')
        rmse, mae = calculate_regression_metrics(y_test, pred_reg)
        reg_summary['LSTM'].append((rmse, mae))
        final_reg_preds['LSTM'] = pred_reg
        print(f"LSTM Regression - RMSE: {rmse:.4f}, MAE: {mae:.4f}")
        plt.figure(figsize=(12,6))
        plt.plot(scaler_target.inverse_transform(y_test)[:1000], label='Actual Mid Prices')
        plt.plot(scaler_target.inverse_transform(pred_reg)[:1000], label='LSTM Prediction')
        plt.xlabel("Time Step")
        plt.ylabel("Mid-Price")
        plt.title(f"Fold {fold_no} - LSTM Regression")
        plt.legend()
        plt.show()
        model_cls = create_lstm_classification_model(X_train.shape)
        model_cls.fit(X_train_lstm, to_categorical(y_train_class_lstm, num_classes=3), epochs=50, batch_size=256,
                      validation_data=(X_test_lstm, to_categorical(y_test_class_lstm, num_classes=3)),
                      callbacks=[early_stop], verbose=0)
        pred_cls = model_cls.predict(X_test_lstm)
        pred_cls = np.pad(pred_cls, ((look_back,0),(0,0)), mode='edge')
        final_class_preds['LSTM'] = pred_cls
        y_pred_cls = np.argmax(pred_cls[look_back:], axis=1)
        acc = accuracy_score(y_test_class_lstm, y_pred_cls)
        class_summary['LSTM'].append(acc)
        print(f"LSTM Classification - Accuracy: {acc:.4f}")
        print(classification_report(y_test_class_lstm, y_pred_cls, zero_division=0))
        compute_metrics(y_test_class_lstm, y_pred_cls, "LSTM")
        current_cm = confusion_matrix(y_test_class_lstm, y_pred_cls)
        if acc > best_lstm_acc:
            best_lstm_acc = acc
            best_lstm_cm = current_cm
        
        # ---- CNN Model ----
        print("\n--- CNN ---")
        model_reg = create_cnn_model(X_train.shape, look_back)
        model_reg.fit(X_train_lstm, y_train_lstm, epochs=10, batch_size=1024,
                      validation_data=(X_test_lstm, y_test_lstm), callbacks=[early_stop], verbose=0)
        pred_reg = model_reg.predict(X_test_lstm)
        pred_reg = np.pad(pred_reg, ((look_back,0),(0,0)), mode='edge')
        rmse, mae = calculate_regression_metrics(y_test, pred_reg)
        reg_summary['CNN'].append((rmse, mae))
        final_reg_preds['CNN'] = pred_reg
        print(f"CNN Regression - RMSE: {rmse:.4f}, MAE: {mae:.4f}")
        plt.figure(figsize=(12,6))
        plt.plot(scaler_target.inverse_transform(y_test)[:1000], label='Actual Mid Prices')
        plt.plot(scaler_target.inverse_transform(pred_reg)[:1000], label='CNN Prediction')
        plt.xlabel("Time Step")
        plt.ylabel("Mid-Price")
        plt.title(f"Fold {fold_no} - CNN Regression")
        plt.legend()
        plt.show()
        model_cls = create_cnn_classification_model(X_train.shape, look_back)
        model_cls.fit(X_train_lstm, to_categorical(y_train_class_lstm, num_classes=3), epochs=50, batch_size=256,
                      validation_data=(X_test_lstm, to_categorical(y_test_class_lstm, num_classes=3)),
                      callbacks=[early_stop], verbose=0)
        pred_cls = model_cls.predict(X_test_lstm)
        pred_cls = np.pad(pred_cls, ((look_back,0),(0,0)), mode='edge')
        final_class_preds['CNN'] = pred_cls
        y_pred_cls = np.argmax(pred_cls[look_back:], axis=1)
        acc = accuracy_score(y_test_class_lstm, y_pred_cls)
        class_summary['CNN'].append(acc)
        print(f"CNN Classification - Accuracy: {acc:.4f}")
        print(classification_report(y_test_class_lstm, y_pred_cls, zero_division=0))
        compute_metrics(y_test_class_lstm, y_pred_cls, "CNN")
        final_class_cm['CNN'] = confusion_matrix(y_test_class_lstm, y_pred_cls)
        
        # ---- Dual-Task Model ----
        print("\n--- Dual-Task ---")
        model_dual = create_dual_task_model(X_train.shape, look_back, encoding_dim=64)
        model_dual.fit(X_train_lstm, {'regression': y_train_lstm, 'classification': to_categorical(create_classification_labels(y_train)[look_back:], num_classes=3)},
                       epochs=10, batch_size=1024,
                       validation_data=(X_test_lstm, {'regression': y_test_lstm, 'classification': to_categorical(create_classification_labels(y_test)[look_back:], num_classes=3)}),
                       callbacks=[early_stop], verbose=0)
        dual_preds = model_dual.predict(X_test_lstm)
        pred_reg = dual_preds[0]
        pred_reg = np.pad(pred_reg, ((look_back,0),(0,0)), mode='edge')
        rmse, mae = calculate_regression_metrics(y_test, pred_reg)
        reg_summary['Dual-Task'].append((rmse, mae))
        final_reg_preds['Dual-Task'] = pred_reg
        print(f"Dual-Task Regression - RMSE: {rmse:.4f}, MAE: {mae:.4f}")
        plt.figure(figsize=(12,6))
        plt.plot(scaler_target.inverse_transform(y_test)[:1000], label='Actual Mid Prices')
        plt.plot(scaler_target.inverse_transform(pred_reg)[:1000], label='Dual-Task Prediction')
        plt.xlabel("Time Step")
        plt.ylabel("Mid-Price")
        plt.title(f"Fold {fold_no} - Dual-Task Regression")
        plt.legend()
        plt.show()
        pred_cls = dual_preds[1]
        pred_cls = np.pad(pred_cls, ((look_back,0),(0,0)), mode='edge')
        final_class_preds['Dual-Task'] = pred_cls
        y_pred_cls = np.argmax(pred_cls[look_back:], axis=1)
        acc = accuracy_score(y_test_class_lstm, y_pred_cls)
        class_summary['Dual-Task'].append(acc)
        print(f"Dual-Task Classification - Accuracy: {acc:.4f}")
        print(classification_report(y_test_class_lstm, y_pred_cls, zero_division=0))
        compute_metrics(y_test_class_lstm, y_pred_cls, "Dual-Task")
        final_class_cm['Dual-Task'] = confusion_matrix(y_test_class_lstm, y_pred_cls)
        
        fold_no += 1
        print(62 * '#')
    
    # ----- AFTER ALL FOLDS: Print Summary -----
    print("\n=== Summary of Regression Results (Average RMSE and MAE) ===")
    avg_reg_metrics = {}
    for model_name, metrics_list in reg_summary.items():
        avg_rmse = np.mean([m[0] for m in metrics_list])
        avg_mae = np.mean([m[1] for m in metrics_list])
        avg_reg_metrics[model_name] = (avg_rmse, avg_mae)
        print(f"{model_name}: RMSE = {avg_rmse:.4f}, MAE = {avg_mae:.4f}")
    
    print("\n=== Summary of Classification Results (Average Accuracy) ===")
    avg_class_accuracy = {}
    for model_name, acc_list in class_summary.items():
        avg_acc = np.mean(acc_list)
        avg_class_accuracy[model_name] = avg_acc
        print(f"{model_name}: Accuracy = {avg_acc:.4f}")
    
    # ----- Plot combined predicted mid-prices for all regression models vs. actual mid-prices -----
    plt.figure(figsize=(12,6))
    plt.plot(scaler_target.inverse_transform(y_test)[:1000], label='Actual Mid Prices', linewidth=2, color='black')
    for model_name in final_reg_preds:
        plt.plot(scaler_target.inverse_transform(final_reg_preds[model_name])[:1000], label=f'{model_name} Prediction')
    plt.xlabel("Time Step")
    plt.ylabel("Mid-Price")
    plt.title("Predicted Mid-Prices of All Regression Models")
    plt.legend()
    plt.show()
    
    # ----- Plot Ensemble Regression Prediction (Average of all regression models) -----
    ensemble_reg_pred = np.mean(np.array(list(final_reg_preds.values())), axis=0)
    ensemble_rmse, ensemble_mae = calculate_regression_metrics(y_test, ensemble_reg_pred)
    print(f"\nEnsemble Regression - RMSE: {ensemble_rmse:.4f}, MAE: {ensemble_mae:.4f}")
    plt.figure(figsize=(12,6))
    plt.plot(scaler_target.inverse_transform(y_test)[:1000], label='Actual Mid Prices', linewidth=2, color='black')
    plt.plot(scaler_target.inverse_transform(ensemble_reg_pred)[:1000], label='Ensemble Prediction')
    plt.xlabel("Time Step")
    plt.ylabel("Mid-Price")
    plt.title("Ensemble Regression Prediction")
    plt.legend()
    plt.show()
    
    # ----- Determine best classification model (using all models) and plot its confusion matrix -----
    avg_cls = avg_class_accuracy  
    best_cls_model = max(avg_cls.items(), key=lambda x: x[1])[0]
    print(f"\nBest Classification Model: {best_cls_model} with Accuracy = {avg_cls[best_cls_model]:.4f}")
    if best_cls_model == 'LSTM':
        best_cm = final_class_cm.get('LSTM', best_lstm_cm)
    else:
        best_cm = final_class_cm[best_cls_model]
    plt.figure(figsize=(6,5))
    sns.heatmap(best_cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Best Classification Model Confusion Matrix: {best_cls_model}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
    
    # ----- Ensemble Classification using Majority Voting -----
    ensemble_votes = []
    for key, pred_prob in final_class_preds.items():
        # Each model's predicted class labels
        ensemble_votes.append(np.argmax(pred_prob, axis=1))
    ensemble_votes = np.array(ensemble_votes)
    # Majority vote along axis 0
    ensemble_class_pred, _ = mode(ensemble_votes, axis=0)
    ensemble_class_pred = ensemble_class_pred.flatten()
    ensemble_class_acc = accuracy_score(y_test_class, ensemble_class_pred)
    print(f"\nEnsemble Classification (Majority Voting) - Accuracy: {ensemble_class_acc:.4f}")
    print(classification_report(y_test_class, ensemble_class_pred, zero_division=0))
    ensemble_class_cm = confusion_matrix(y_test_class, ensemble_class_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(ensemble_class_cm, annot=True, fmt='d', cmap='Blues')
    plt.title("Ensemble Classification Confusion Matrix (Majority Voting)")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
    
    # ----- Plot summary bar charts for classification macro metrics -----
    classifier_names = list(class_metrics.keys())
    precisions = [class_metrics[name][0] for name in classifier_names]
    recalls = [class_metrics[name][1] for name in classifier_names]
    f1_scores = [class_metrics[name][2] for name in classifier_names]
    
    plt.figure(figsize=(10,6))
    plt.bar(classifier_names, precisions, color='skyblue')
    plt.title("Macro Precision per Classifier")
    plt.ylabel("Precision")
    plt.ylim(0,1)
    plt.show()
    
    plt.figure(figsize=(10,6))
    plt.bar(classifier_names, recalls, color='lightgreen')
    plt.title("Macro Recall per Classifier")
    plt.ylabel("Recall")
    plt.ylim(0,1)
    plt.show()
    
    plt.figure(figsize=(10,6))
    plt.bar(classifier_names, f1_scores, color='salmon')
    plt.title("Macro F1-Score per Classifier")
    plt.ylabel("F1-Score")
    plt.ylim(0,1)
    plt.show()

# Run the entire protocol
run_models_simultaneously()
```
